<h1>Missing Data Handing:</h1>
<b>1. Find missing values in the dataset:</b><br>
The <b>isnull( )</b> detects the missing values and returns a boolean object indicating if the values are NA. 

```python
#find missing values
print(data_frame.isnull())
```

<b>2. Find the number of missing values in the dataset:</b><br>
To find out the number of missing values in the dataset, use <b>data_frame.isnull( ).sum( )</b>.

```python
# total missing values
print(data_frame.isnull().sum())
```

<b>3. Removing missing values:</b><br>
The <b>data_frame.dropna( )</b> function removes columns or rows which contains atleast one missing values.<br>
By default, data_frame.dropna( ) <b>drops the rows where at least one element is missing</b>.<br>
<b>data_frame.dropna(axis = 1) drops the columns where at least one element is missing.</b>

```python
# remove missing values
print(data_frame.dropna())
```

<b>4. Fill in missing values:</b><br>
We can fill null values using <b>data_frame.fillna( )</b> function.<br>
  <b>data_frame = data_frame.fillna(value)</b><br>
But by using the above format all the null values will get filled with the same values. To fill different values in the different columns we can use.<br>
  <b>data_frame[col] = data_frame[col].fillna(value)</b>

<h1>Sorting DataFrame values:</h1>
<b>1. Sort by column:</b><br>
The <b>sort_values( )</b> are the values of the column whose name is passed in the <b>by</b> attribute in the ascending order by default<br> we can set this attribute to false to sort the array in the descending order.

```python
# sort column by desending order
data_frame.sort_values(by="Salary",ascending=False).head()
```

<b>2.  Sort by multiple columns:</b>

<h1>Merge Data Frames:</h1>
The <b>merge()</b> function in pandas is used for all standard database join operations.<br>
Merge operation on data frames will join two data frames based on their common column values.

```python
#Creating dataframe1
df1 = pd.DataFrame({
        'Name':['Jeevan', 'Raavan', 'Geeta', 'Bheem'], 
        'Age':[25, 24, 52, 40], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']})
df1
```

```python
#Creating dataframe2
df2 = pd.DataFrame({'Name':['Jeevan', 'Raavan', 'Geeta', 'Bheem'],
                    'Salary':[100000, 50000, 20000, 40000]})
df2
```

<b>Now. let's merge these two data frames created earlier.</b>

```python
#Merging two dataframes
df = pd.merge(df1, df2)
df
```

<h1>Apply Function:</h1>
<b>1. By defining a function beforehand:</b><br>
The <b>apply( )</b> function is used to iterate over a data frame. It can also be used with lambda functions.

```python
# apply function happy
def happy(value):
    if value > 10.00:
        return "Happy"
    else:
        return "Sad"

# cread new column in data_frame and find who is happy with bonus
data_frame["Satisfaction"]=data_frame["Bonus %"].apply(happy)
display(data_frame.head(10))
```

<b>2. By using the lambda operator:</b><br>
This syntax is generally used to apply log transformations and normalize the data to bring it in the range of 0 to 1 for particular columns of the data.

```python
const = data_frame["Salary"].max()
data_frame["Salary_lamda"]=data_frame["Salary"].apply(lambda x: x/const)
display(data_frame.head(10))
```
