# Steps for Data Normalization in Pandas.
<b>setps :</b><br>
1. Import the required libraries<br>
2. Load or create a dataset<br>
3. Apply different normalization techniques<br>
4. Visualize the results<br>

```python
# creat simple dataset
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame([
    [180000, 110, 18.9, 1400],  
    [360000, 905, 23.4, 1800],  
    [230000, 230, 14.0, 1300],  
    [60000, 450, 13.5, 1500]],columns=["Col A","Col B","Col C","Col D"])
print(df)
df.plot(kind="bar")
plt.show()
```

<h1>Normalization Techniques in Pandas:</h1>
<b>1. Maximum Absolute Scaling:</b>
<p>This technique rescales each feature between -1 and 1 by dividing all values by the maximum absolute value in that column. This technique is especially useful when your data doesn’t contain negative numbers and you want to preserve the data’s sparsity. We can apply the maximum absolute scaling in Pandas using the <b>.max() and .abs()</b></p>
<b>syntax:<br>
    for column in max_scal.columns:<br>
    max_scal[column]=max_scal[column]/max_scal[column].abs().max()</b>

```python
# for example 
max_scal = df.copy()
for column in max_scal.columns:
    max_scal[column]=max_scal[column]/max_scal[column].abs().max()

print(max_scal)
max_scal.plot(kind="bar")
plt.show()
```

<p>above output all values now lie between -1 and 1. Each value is shown in relation to the largest value in that column.</p>

<b>2. The min-max feature scaling:</b>
<p>The min-max approach also called normalization rescales the feature to a hard and fast range of [0,1] by subtracting the minimum value of the feature then dividing by the range.<br>
can apply the min-max scaling in Pandas using the <b>.min() and .max()</b> methods.</p><br>
<b>Syntax:<br>
     for column in df.columns:<br>
         df[column] = (df[column]-df[column].min())/(df[column].max()-df[column].min())</b>

```python
min_max = df.copy()
for column in min_max.columns:
    min_max[column] = (min_max[column]-min_max[column].min())/(min_max[column].max()-min_max[column].min())

print(min_max)
min_max.plot(kind="bar")
plt.show()
```

<b>3. The z-score method:</b><br>
<P>The z-score method often called standardization changes the values in each column so that they have a mean of 0 and a standard deviation of 1. This technique is best when your data follow a normal distribution or when you want to treat values in terms of how far they are from the average.</P>


```python
# creat simple dataset
import pandas as pd
import matplotlib.pyplot as plt
df2 = pd.DataFrame([
    [180000, 110, 18.9, 1400],  
    [360000, 905, 23.4, 1800],  
    [230000, 230, 14.0, 1300],  
    [60000, 450, 13.5, 1500]],columns=["Col A","Col B","Col C","Col D"])
print(df2)
df2.plot(kind="bar")
plt.show()
```

```python
# example: Z score use:
z_scaled = df2.copy()

for column in z_scaled.columns:
    z_scaled[column]=(z_scaled[column]-z_scaled[column].mean())/z_scaled[column].std()
print(z_scaled)
z_scaled.plot(kind="bar")
plt.show()
```
